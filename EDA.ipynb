```python
# Se importa el dataset edx_courses.csv.
edx_courses = pd.read_csv('Datasets/edx_courses.csv')

# Se eliminan valores nulos en la cantidad de personas que inscritas.
edx_courses = edx_courses.dropna(subset=["n_enrolled"])

# Se eliminan las comas y se transforma esta variable a int64.
edx_courses.n_enrolled = edx_courses.n_enrolled.str.replace(',', '')
edx_courses.n_enrolled = edx_courses.n_enrolled.astype('int64')

# Se eliminan duplicados del dataframe.
edx_courses.drop_duplicates(inplace=True)
edx_courses.head(3)
```

```python
# AnÃ¡lisis del dataset
edx_report = sv.analyze(edx_courses)
edx_report.show_notebook()
```

La gente que se inscribe a un curso estÃ¡ fuertemente correlacionada a las horas de estudio por semana. (0.73)

La gente que se inscribe a un curso estÃ¡ levemente correlacionada al precio del curso. (0.47)


```python
levels = edx_courses[['Level', 'n_enrolled']].groupby(by='Level').sum()
avanzado = levels.iloc[0]/levels.sum()
intermedio = levels.iloc[1]/levels.sum()
introductorio = levels.iloc[2]/levels.sum()

print(f"El {round(introductorio[0]*100,2)}" + "%" + " de las personas se inscriben a cursos introductorios.")
print(f"El {round(intermedio[0]*100,2)}" + "%" + " de las personas se inscriben a cursos intermedios.")
print(f"El {round(avanzado[0]*100,2)}" + "%" + " de las personas se inscriben a cursos avanzados.")
```
